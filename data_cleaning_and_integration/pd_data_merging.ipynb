```python
# Main PD file
main_pd_df = pd.read_csv("../data/pd_calls_for_service_2022_datasd.csv")
```

```python
# neighborhood beats 
nh_beats_df = pd.read_csv("../data/pd_beat_codes_list_datasd.csv")
nh_beats_df.head()
```

```python
# merge pd file with beats file for neighbourhoods
main_pd_df = main_pd_df.merge(nh_beats_df,on='beat',how='left')
print("count of main_pd_df",len(main_pd_df))
main_pd_df.head()
```

```python
# Crime Types for the codes
crime_code_type_df = pd.read_csv("../data/pd_cfs_calltypes_datasd.csv")
crime_code_type_df.head()
```

```python
len(crime_code_type_df)
```

```python
# Checking the unique count of call_type
print("Checking the unique count of call_type",len(pd.unique(crime_code_type_df['call_type'])))
```

```python
# Total count of the crime call type file.
print("Total count of the crime call type file.",len(crime_code_type_df))
```

```python
# Selecting duplicate rows except first
# occurrence based on all columns
duplicate = crime_code_type_df[crime_code_type_df['call_type'].duplicated()]
 
print("Duplicate Rows :")
 
# Print the resultant Dataframe
duplicate
```

```python
# Removing duplicates 
crime_code_type_df = crime_code_type_df[~crime_code_type_df['call_type'].duplicated()]
```

```python
# Checking the unique count of call_type
print("Checking the unique count of call_type",len(pd.unique(crime_code_type_df['call_type'])))
print("Total count of the crime call type file.",len(crime_code_type_df))
```

```python
# merge pd file with crime_type file
main_pd_df = main_pd_df.merge(crime_code_type_df,on='call_type',how='left')

# Renaming description column to call_type_desc
main_pd_df.rename(columns = {'description':'call_type_desc'}, inplace = True)

print("count of main_pd_df",len(main_pd_df))
main_pd_df.head()
```

```python
main_pd_df.to_csv("../data/processed_pd_data/merged_pd_data.csv",header=True)
```

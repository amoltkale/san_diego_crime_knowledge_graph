```python
%%time
# Regex patten to extract location and date from the merged column
pattern = re.compile(r"(.*\/?.*)\/(Edited )?(\d{1,2}[ -][a-zA-Z]{3}[ -]?[0-9]{0,2})")
for index, row in next_door_df.iterrows():
    location_date = str(row['Location/Date'])
    #print(location_date)
    try:
        
        groups = list(pattern.finditer(location_date))[0]
        next_door_df.at[index,'Neighborhood'] = groups.group(1)
        next_door_df.at[index,'Date'] = groups.group(3)

    except IndexError:
        next_door_df.at[index,'Date'] = location_date
```

```python
def validate_date(date_text,date_format):
    '''This function validates if a date is following a given format.'''
    try:
        datetime.strptime(date_text, date_format)
        return True
    except ValueError:
        # raise ValueError("Incorrect data format, should be YYYY-MM-DD")
        return False
```

```python
def get_proper_date(date_text):
    '''This validates multiple data formats I have observed in our data and converts it into uniform date'''
    if validate_date(date_text,'%d %b'):
        return datetime.strptime(date_text, '%d %b').replace(year=2022).strftime('%m-%d-%Y')
    elif validate_date(date_text,'%d-%b'):
        return datetime.strptime(date_text, '%d-%b').replace(year=2022).strftime('%m-%d-%Y')
    elif validate_date(date_text,'%d-%b-%y'):
        return datetime.strptime(date_text, '%d-%b-%y').strftime('%m-%d-%Y')
    elif validate_date(date_text,'%d %b %y'):
        return datetime.strptime(date_text, '%d %b %y').strftime('%m-%d-%Y')
    elif validate_date(date_text,'%d %b %Y'):
        return datetime.strptime(date_text, '%d %b %Y').strftime('%m-%d-%Y')
    elif validate_date(date_text,'%d-%b-%Y'):
        return datetime.strptime(date_text, '%d-%b-%Y').strftime('%m-%d-%Y')
```

```python
next_door_df['Date'] =  next_door_df['Date'].apply(lambda x: get_proper_date(x))
print("count of next_door_df",len(next_door_df))
next_door_df.head()
```

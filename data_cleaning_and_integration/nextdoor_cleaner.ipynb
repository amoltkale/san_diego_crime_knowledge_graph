```python
import pandas as pd
import re
from cleaner_lib import clean_str_col, translate_emojis
```

```python
next_door_df = pd.read_csv("../data/raw_nextdoor/SanDiegoCrimesNextDoor.csv")
print("count of next_door_df",len(next_door_df))
next_door_df.head()
```

```python
%%time
# Regex patten to extract location and date from the merged column
pattern = re.compile(r"(.*\/?.*)\/(Edited )?(\d{1,2}[ -][a-zA-Z]{3}[ -]?[0-9]{0,2})")
for index, row in next_door_df.iterrows():
    location_date = str(row['Location/Date'])
    #print(location_date)
    try:
        
        groups = list(pattern.finditer(location_date))[0]
        next_door_df.at[index,'Location'] = groups.group(1)
        next_door_df.at[index,'Date'] = groups.group(3)

    except IndexError:
        next_door_df.at[index,'Date'] = location_date
```

```python
print("count of next_door_df",len(next_door_df))
next_door_df.head()
```

```python
next_door_df = translate_emojis(next_door_df,'TopComments')
print("count of next_door_df",len(next_door_df))
next_door_df.head()
```
